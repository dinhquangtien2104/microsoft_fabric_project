```python
path= "abfss://4155d96d-1328-43ad-982f-7aecb1ee2008@onelake.dfs.fabric.microsoft.com/900594c3-14cd-4919-8406-174ac04587bb/Files/"
shipping_mode = spark.read.option("header", "true") .csv(path + "shipping_mode")
transaction = spark.read.option("header", "true") .csv(path + "transaction")
order_status = spark.read.option("header", "true") .csv(path + "order_status")
delivery_status  = spark.read.option("header", "true") .csv(path + "delivery_status")
delivery_risk = spark.read.option("header", "true") .csv(path + "delivery_risk")

```

```python
from pyspark.sql.functions import current_timestamp, lit, row_number,col,when
from pyspark.sql.window import Window
from functools import reduce
from pyspark.sql import functions as F
```

# Create dim_shipping_mode

```python

# Define a window specification
window_spec_key = Window.orderBy("shipping_mode")  
# Add a new column 'shipping_mode_key' with consecutive IDs starting from 1
shipping_mode = shipping_mode.withColumn("shipping_mode_key", row_number().over(window_spec_key))

shipping_mode = shipping_mode.select("shipping_mode_key", "shipping_mode")
display(shipping_mode)

```

# Create dim_transaction

```python

# Define a window specification
window_spec_key = Window.orderBy("type_of_transaction")  
# Add a new column 'transaction' with consecutive IDs starting from 1
transaction = transaction.withColumn("transaction_key", row_number().over(window_spec_key))

transaction = transaction.select("transaction_key", "type_of_transaction")

display(transaction)

```

# Create dim_order_status

```python

# Define a window specification
window_spec_key = Window.orderBy("order_status")  
# Add a new column 'transaction' with consecutive IDs starting from 1
order_status = order_status.withColumn("order_status_key", row_number().over(window_spec_key))

order_status = order_status.select("order_status_key", "order_status")

display(order_status)

```

# Create dim_delivery_status

```python

# Define a window specification
window_spec_key = Window.orderBy("delivery_status")  
# Add a new column 'transaction' with consecutive IDs starting from 1
delivery_status = delivery_status.withColumn("delivery_status_key", row_number().over(window_spec_key))

delivery_status = delivery_status.select("delivery_status_key", "delivery_status")

display(delivery_status)

```

# Create dim_delivery_risk

```python

# Define a window specification
window_spec_key = Window.orderBy("late_delivery_risk")  
# Add a new column 'transaction' with consecutive IDs starting from 1
delivery_risk = delivery_risk.withColumn("delivery_risk_key", row_number().over(window_spec_key))

delivery_risk = delivery_risk.select("delivery_risk_key", "late_delivery_risk")


delivery_risk = delivery_risk.withColumn("late_delivery_risk" , when(col("late_delivery_risk") == '0', "late").otherwise("not late"))
display(delivery_risk)

```
